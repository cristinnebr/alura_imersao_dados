```python
# Explorar dados de anúncios de imóveis em São Paulo
import pandas as pd
```

```python
# Carregar dados
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url)
```

```python
# Selecionar os imóveis em Pinheiros
tem_imoveis_pinheiros = ( dados["Bairro"] == "Pinheiros")
imoveis_pinheiros = dados[tem_imoveis_pinheiros]
imoveis_pinheiros
```

```python
# Tarefa 5: Que bairros tem rua sem nome?
# Criando subset apenas com ruas sem nome
na_rua = ( dados["Rua"].isna())
rua_sem_nome = dados[na_rua]
rua_sem_nome
```

```python
# Mostrar os bairros do subset rua_sem_nome
rua_sem_nome["Bairro"]
```

```python
#Tarefa 3: Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão
n_quartos = dados["Quartos"].value_counts() # a maioria dos imóveis tem 3 quartos, e em segundo lugar 4 quartos
n_quartos.plot.bar()
```

```python
metragem_maior_100 = dados[dados["Metragem"] > 100] # seleção de imóveis com metragem acima de 100
metragem_maior_100
```

```python
n_metragem_bairro = metragem_maior_100.Bairro.value_counts() # quantidade de imóvies por bairro com metragem acima de 100
n_metragem_bairro.head(20).plot.bar()
```

```python
metragem_menor100 = dados[dados["Metragem"] < 100] # seleção de imóveis com metragem abaixo de 100
metragem_menor100
```

```python
n_metragem_bairro_men100 = metragem_menor100.Bairro.value_counts() # # quantidade de imóvies por bairro com metragem abaixo de 100
n_metragem_bairro_men100.head(20).plot.bar()
```

```python
#Tarefa 1: Realizar a média da metragem para cara um dos bairros
# com .pivot_table: mais complexo
metragem_bairro = dados[["Bairro", "Metragem"]]
metragem_bairro
```

```python
mean_metragem_bairro = dados.pivot_table(values = "Metragem", index = "Bairro", 
                  aggfunc = 'mean')
mean_metragem_bairro
```

```python
# com groupby e mean: mais simples
dados.groupby('Bairro').mean()['Metragem']
```
